```
#import 
import numpy as np
import pandas as pd 
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential as seq
from keras.layers import Dense, Flatten, Conv2D, Dropout
from keras.layers import MaxPooling2D as mp2d 
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import MinMaxScaler as MMS
from sklearn.metrics import confusion_matrix as CM
```

```
data = pd.read_csv('data.csv')
data = data.dropna(axis=0)
data.shape
```

```
#visualize
data.head(5)
```

```
#scale data
mms = MMS()
Y = data['Bankrupt?'].values
data = data.drop(columns=['Bankrupt?'])
data = mms.fit_transform(data)
```

```
#split
trainX,testX,trainY,testY = tts(data, Y, test_size=0.2, random_state=42)
```

```
#create architecture
arch = seq()
arch.add(Dense(units = 96, input_shape= (95,), activation = 'relu')) #first dense layer
arch.add(Dense(units = 64, activation = 'relu')) #second dense layer
arch.add(Dense(units = 32, activation = 'relu')) #second dense layer
arch.add(Dense(units = 2, activation = 'softmax')) #last hidden layer
arch.compile(optimizer= 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```
arch.fit(trainX,trainY,batch_size=16,epochs=8)
```

```
predictions = arch.predict(testX)
decisions = predictions.argmax(axis=1)
cm = CM(decisions,testY)
print("ANN accuracy: ", (cm[0][0] + cm[1][1])/(cm[0][0] + cm[1][1] + cm[1][0] + cm[0][1]) *100, '%')
```
